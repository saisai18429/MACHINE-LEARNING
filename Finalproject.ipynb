Mobile phone price fixed on the various factors like processor, RAM, Internal memory etc., There are 21 features given in the dataset. Classify the mobile price range as 0,1,2 or 3 based upon the features given in the dataset.

Class 0 – low cost

Class 1 – medium cost

Class 2 -  high cost

Class 3 – very high cost

#Use SVM algorithm for classification.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mtb
import seaborn as sns
```

```python
df = pd.read_csv("/content/mobile data set - mobile data set.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
X = df.drop("price_range",axis=1)
y =df["price_range"]
```

```python
print(X)
```

```python
print(y)
```

# Train And Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

#Training the SVM Model

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear', C=1.0)
```

```python
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
values = [[1512,0,0.9,0,4,1,46,0.1,145,5,5,336,670,869,18,10,19,1,1,1]]
new_pred = svm_model.predict(values)
if new_pred == 0:
  print("low cost")
elif new_pred == 1:
  print("medium cost")
elif new_pred == 2:
  print("high cost")
elif new_pred == 3:
  print("very high_cost")
```
